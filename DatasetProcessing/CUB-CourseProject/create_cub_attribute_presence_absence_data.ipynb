```python
img_attr_list = []
with open('/data/CUB/CUB-Original/CUB_200_2011/attributes/image_attribute_labels.txt', 'r') as f:
    data = f.readlines()
```

```python
for d in tqdm(data):
    d = d.strip().replace('  ', ' ')
    img_attr_val = d.split()[:3]
    img_attr_list.append(dict(zip(['image_id', 'attribute_id', 'value'], img_attr_val)))
```

```python
img_attr_df = pd.DataFrame(img_attr_list)
```

```python
img_attr_pivot_df = img_attr_df.pivot(index='image_id', columns='attribute_id', values='value')
```

```python
img_attr_pivot_df = img_attr_pivot_df.reset_index()
```

```python
# Read in the attributes
with open('/data/CUB/CUB-Original/CUB_200_2011/attributes/attributes.txt', 'r') as f:
    attributes = f.readlines()
```

```python
attr_id_name_dict = {}
for attr in tqdm(attributes):
    attr_id, attr_name = attr.strip().split()
    attr_id_name_dict[attr_id] = attr_name
    
```

```python
img_attr_pivot_df = img_attr_pivot_df.rename(columns=attr_id_name_dict)
```

```python
# Read in the image_id vs image_path file
with open('/data/CUB/CUB-Original/CUB_200_2011/images.txt', 'r') as f:
    images = f.readlines()
```

```python
image_id_path_dict = {}
for img in tqdm(images):
    image_id, path = img.strip().split()
    image_id_path_dict[image_id] = path
```

```python
img_attr_pivot_df['image_path'] = img_attr_pivot_df['image_id'].map(image_id_path_dict)
```

# Read CSV file and perform further analysis/processing

```python
import pandas as pd
```

```python
cub_attr = pd.read_csv('cub_attribute_classification.csv')
```

```python
list(cub_attr.columns)
```

```python
cub_attr['index_id'] = cub_attr.index
```

```python
# Select 10 attributes manually to create a smaller dataset
selected_attributes = [
    'has_wing_color::blue',
    'has_head_pattern::eyeline',
    'has_breast_color::yellow',
    'has_throat_color::yellow',
    'has_eye_color::grey',
    'has_forehead_color::blue',
    'has_under_tail_color::black',
    'has_nape_color::red',
    'has_wing_shape::rounded-wings',
    'has_tail_pattern::striped',   
]
```

```python
cub_attr_small = cub_attr[selected_attributes + ['image_id', 'image_path', 'index_id']]
```

```python
len(cub_attr_small)
```

```python
sum(cub_attr_small['has_wing_color::blue'] == 1)
```

```python
# For each trait, keep all of the ones, and an equal amount of non present ones
# Drop duplicates
# Whatever imbalance exists, just keep it, since you will be using ASLLoss anyways
```

```python
req_df = pd.DataFrame()
for attr in selected_attributes:
    req_df = pd.concat([req_df, cub_attr_small[cub_attr_small[attr] == 1]])
```

```python
req_df = req_df.drop_duplicates('index_id')
```

```python
len(req_df)
```

```python
attr_presence_counts = {}
for attr in selected_attributes:
    attr_presence_counts[attr] = sum(req_df[attr] == 1)
```

```python
attr_presence_counts
```

```python
req_df = req_df.drop(columns=['index_id'])
```

```python
req_df.to_csv('cub_attr_small.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split 
```

```python
from tqdm import tqdm
# req_df_strat = pd.DataFrame()
# for i, row in tqdm(req_df.iterrows()):
#    row['stratify'] = ''.join([str(row[a]) for a in selected_attributes])
#    req_df_strat = pd.concat([req_df_strat, row])
```

```python
req_df['stratify'] = req_df.apply(lambda row: "".join([ str(row[a]) for a in selected_attributes]), axis=1)
```

```python
vc = req_df.stratify.value_counts()
vc_to_default_train = vc[vc.values <= 5].index
```

```python
default_train = req_df[req_df.stratify.isin(vc_to_default_train)]
```

```python
len(default_train)
```

```python
remaining_df = req_df[~req_df.stratify.isin(vc_to_default_train)]
```

```python
len(remaining_df)
```

```python
train_val, test = train_test_split(remaining_df, test_size=0.1, stratify=remaining_df.stratify)
```

```python
train, val = train_test_split(train_val, test_size=0.1, stratify=train_val.stratify)
```

```python
train = pd.concat([train, default_train])
```
